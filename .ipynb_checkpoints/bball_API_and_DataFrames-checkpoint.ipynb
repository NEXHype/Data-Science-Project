```python
# imports
import requests
import pandas as pd
```

```python
#checking API endpoint for all stats for latest season
response = requests.get('https://www.balldontlie.io/api/v1/stats?seasons[]=2022')
```

```python
response_json = response.json()
response_json
```

```python
#extracting all relevant data from API, API had a 60 calls per minute limit, and since we needed to call the API for each page of info,
#60 pages were called each time until all 438 pages have been looped through
id_list = []
fg3a_list = []
fg3m_list = []
fg2a_list = []
fg2m_list = []
page_num = 1

while page_num < 61:
    response = requests.get('https://www.balldontlie.io/api/v1/stats?seasons[]=2022&per_page=100&page=' + str(page_num))
    response_json = response.json()
    for i in response_json['data']:
        id_list.append(i['player']['id'])
        fg3a_list.append(i['fg3a'])
        fg3m_list.append(i['fg3m'])
        fg2a_list.append((i['fga'] - i['fg3a']))
        fg2m_list.append((i['fgm'] - i['fg3m']))
    page_num += 1
        
```

```python
#creating dataframe dict
df1_dict = {'id': id_list, 'fg3m': fg3m_list, 'fg3a': fg3a_list, 'fg2m': fg2m_list, 'fg2a': fg2a_list}
```

```python
initial_df = pd.DataFrame(df1_dict)
initial_df
```

```python
#grouping by player id and summing all the shooting numbers
initial_df_grouped = initial_df.groupby('id').sum()
```

```python
initial_df_grouped
```

```python
#generating shooting percentages for each player by looping through dataframe
three_percent_list = []
for ind in initial_df_grouped.index:
    try:
        three_percent_list.append(round(((initial_df_grouped['fg3m'][ind] / initial_df_grouped['fg3a'][ind])) * 100, 1))
    except:
        three_percent_list.append('NaN')
```

```python
three_percent_list
```

```python
two_percent_list = []
for ind in initial_df_grouped.index:
    try:
        two_percent_list.append(round(((initial_df_grouped['fg2m'][ind] / initial_df_grouped['fg2a'][ind])) * 100, 1))
    except:
        two_percent_list.append('NaN')
```

```python
two_percent_list
```

```python
initial_df_grouped['3pt_percent'] = three_percent_list
```

```python
initial_df_grouped['2pt_percent'] = two_percent_list
```

```python
#checking API endpoint for all player info 
response2 = requests.get('https://www.balldontlie.io/api/v1/players')
response2_json = response2.json()
```

```python
response2_json['data']
```

```python
#extracting all relevant data from API, and changing the position letters to their corresponding int value for future use
page_num_v2 = 1
id_list_v2 = []
name_first = []
name_last = []
height_inch = []
weight_lb = []
position_list = []
while page_num_v2 < 53:
    response2 = requests.get('https://www.balldontlie.io/api/v1/players?per_page=100&page=' + str(page_num_v2))
    response2_json = response2.json()
    for i in response2_json['data']:
        if i['height_feet'] == None:
            pass
        else:
            id_list_v2.append(i['id'])
            name_first.append(i['first_name'])
            name_last.append(i['last_name'])
            height_inch.append((i['height_feet']*12 + i['height_inches']))
            weight_lb.append((i['weight_pounds']))
            if i['position'] == '':
                position_list.append('NaN')
            elif i['position'][0] == 'G':
                position_list.append(1)
            elif i['position'][0] == 'F':
                position_list.append(3)
            elif i['position'][0] == 'C':
                position_list.append(5)
            else:
                position_list.append('NaN')
    page_num_v2 += 1
```

```python
page_num_v2
```

```python
#generating dataframe dict for player info
df2_dict = {'id': id_list_v2, 'first_name': name_first, 'last_name': name_last, 'height_inches':height_inch, 'weight_pounds': weight_lb, 'position': position_list}
```

```python
df2 = pd.DataFrame(df2_dict)
df2
```

```python
#joining both dataframes together to get resulting final dataframe will all shooting stats and player info
df_final_join = pd.merge(df1, df2, on='id', how='inner')
df_final_join
```

```python
df_final_join.to_csv('joined_data_nba.csv')
```
